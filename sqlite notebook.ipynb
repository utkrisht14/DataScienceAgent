```python
## SQLite DB - Storing and Retrieving Vectors
```

```python
import sqlite3
import numpy as np
```

```python
# Create a connection to the SQLite DB
conn = sqlite3.connect("test.db")
```

```python
# Create a cursor object to execute different sql element
cursor = conn.cursor()
```

```python
# Create a table for vector data
cursor.execute(
    """
    CREATE TABLE IF NOT EXISTS vectors (
        id INTEGER PRIMARY KEY,
        vector BLOB  NOT NULL

    );
    """
)
```

```python
# Generate some vectors using numpy
vect1 = np.array([1.2, 3.4, 2.1, 0.8])
vect2 = np.array([2.7, 1.5, 3.9, 2.3])
```

```python
vect1.tobytes()
```

```python
cursor.execute(
    """
    INSERT INTO vectors (vector) VALUES (?)
    """,
    (sqlite3.Binary(vect1.tobytes()),)
)
```

```python
cursor.execute(
    """
    INSERT INTO vectors (vector) VALUES (?)
    """,
    (sqlite3.Binary(vect2.tobytes()),)
)
```

```python
# Retrieve the data from the vectors
cursor.execute("SELECT * FROM vectors")
```
